```python
#To import required libraries to solve q2
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import imutils
%matplotlib inline
```

```python
def computeImageFeatures(image, method):
    if method == "HoG":
        hog = cv2.HOGDescriptor()
        h = hog.compute(image, None)
        
        print("HoG is computed")
        
        return h
        
    elif method == "Dense":
        sift = cv2.xfeatures2d.SIFT_create()
        step_size = 5
        keyPoints = [cv2.KeyPoint(x, y, step_size) for y in range(0, image.shape[0], step_size) 
                                    for x in range(0, image.shape[1], step_size)]
        dense = sift.compute(image, keyPoints)
        
        print("Dense is computed")

        return dense
    
    elif method =="SIFT":
        sift = cv2.xfeatures2d.SIFT_create()
        keyPoints, descriptors = sift.detectAndCompute(image, None)
        
        print("SIFT is computed")
  
        return keyPoints, descriptors
```

```python
image = cv2.imread("mp1_1.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imageFeatures = []

methods = ["HoG", "Dense", "SIFT"]

for method in methods:
    imageFeatures.append(computeImageFeatures(image, method))

```
