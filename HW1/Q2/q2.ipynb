```python
#To import required libraries to solve q2
import cv2 
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
```

```python
# To read two images 
image1 = cv2.imread("p1.jpg")
image2 = cv2.imread("p2.jpg")
```

```python
# To show images:
figure, ax = plt.subplots(1, 2, figsize=(16, 16))


ax[0].imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
ax[1].imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
```

```python
#To initiate the SIFT detector:
sift = cv2.xfeatures2d.SIFT_create()


#To compute the keypoints and descriptors for images using SIFT
keyPoints1, descriptors1 = sift.detectAndCompute(image1, None)
keyPoints2, descriptors2 = sift.detectAndCompute(image2, None)
```

```python
#To draw detected keypoints on the first image and show them:
plt.figure(figsize = (13,13))
image1_KD = cv2.drawKeypoints(image1, keyPoints1, image1)
plt.imshow(cv2.cvtColor(image1_KD, cv2.COLOR_BGR2RGB), interpolation='none', aspect='auto')
```

```python
# To draw detected keypoints on image 2 and show them:

plt.figure(figsize = (13,13))
image2_KD = cv2.drawKeypoints(image2, keyPoints2, image2)
plt.imshow(cv2.cvtColor(image2_KD, cv2.COLOR_BGR2RGB), interpolation='none', aspect='auto')
```

```python
#To setup feature matching tool:
sift_match = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = sift_match.match(descriptors1,descriptors2)
matches = sorted(matches, key = lambda x:x.distance)
```

```python
#To draw the matching result:
plt.figure(figsize = (13,13))
finalResult = cv2.drawMatches(image1_KD, keyPoints1, image2_KD, keyPoints2, matches[:50], image2_KD, flags=2)
plt.imshow(cv2.cvtColor(finalResult, cv2.COLOR_BGR2RGB), interpolation='none', aspect='auto'),plt.show()
```
